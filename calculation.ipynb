# Tunable optical absorption in undoped graphene sandwich

The propagation of the light is represented by the matrix equation, which is known as the transfer matrix method. In the transfer matrix method, the $E^{(+)}$ and $E^{(-)}$ are expressed in terms of $E''^{(+)}$ and $E''^{(-)}$. Thus, we obtain the following matrix equation,

\begin{align}
  \begin{bmatrix}
          E^{(+)}\\
          E^{(-)}
  \end{bmatrix} = M_{0|n}P_nM_{n|n'}P_{n'}
  \begin{bmatrix}
          E^{''(+)}\\
          E^{''(-)}
  \end{bmatrix}
  ,
\end{align}

where $P_n$ and $M_{n|n'}$ are the propagation matrix of medium $n$ and matching matrix between medium $n$ and $n'$, respectively. The product of all $P_n$ and $M_{n|n'}$ is known as the transfer matrix of the structure. The $P_n$ and $M_{n|n'}$ are expressed as follows:

\begin{align}
  M
  &=
  \frac{1}{2} \begin{bmatrix}
  \left( 1 + \displaystyle\frac{n'}{n} \right) & \left( 1 - \displaystyle\frac{n'}{n} \right)\\
  \left( 1 - \displaystyle\frac{n'}{n} \right) & \left( 1 + \displaystyle\frac{n'}{n} \right)
  \end{bmatrix}
\end{align}

\begin{align}
P_n =
  \begin{bmatrix}
          - i & 0\\
          0   & i
  \end{bmatrix}\equiv P
\end{align}

In the mirror structures consisting of repeating unit layers of ABC and ABCD, respectively, each layer of the unit layers can take a different refractive index. The transfer matrix of the structure consists of the following matrix equations, with $x$ defining how many media are used in the unit layers:

\begin{align}
  J_1 &= M_{0|1} P ... M_{m-1|m} P, \notag\\
  J_2 &= M_{m|1} P ... M_{m-1|m} P, \notag\\
  J_3 &= P M_{m|m-1} ... P M_{1|m}, \notag\\
  J_4 &= P M_{m|m-1} ... P M_{1|0},
\end{align}

where $0$ denotes air.

The transfer matrices for the ABC unit layer are given as follows:
    
$$\begin{align}
  &J_{1 (3)} = \displaystyle\frac{i}{2}
  \begin{bmatrix}
          \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} + \displaystyle\alpha_{3} & - \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} + \displaystyle\alpha_{3}\\
          \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} - \displaystyle\alpha_{3} & - \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} - \displaystyle\alpha_{3}
  \end{bmatrix},
\end{align}$$

$$\begin{align}
  &J_{2 (3)} = \displaystyle\frac{i}{2}
  \begin{bmatrix}
          \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} + \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} & - \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} + \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}}\\
          \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} - \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} & - \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} - \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}}
  \end{bmatrix},
\end{align}$$

$$\begin{align}
  &J_{3 (3)} = \displaystyle\frac{i}{2}
  \begin{bmatrix}
          \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} + \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} & \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} - \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}}\\
          - \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} + \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}} & - \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} - \displaystyle\frac{n_{C}}{\displaystyle\alpha_{3}}
  \end{bmatrix},
\end{align}$$

$$\begin{align}
  &J_{4 (3)} = \displaystyle\frac{i}{2}
  \begin{bmatrix}
          \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} + \displaystyle\frac{1}{\displaystyle\alpha_{3}} & \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} - \displaystyle\frac{1}{\displaystyle\alpha_{3}}\\
          - \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} + \displaystyle\frac{1}{\displaystyle\alpha_{3}} & - \displaystyle\frac{\displaystyle\alpha_{3}}{n_{C}} - \displaystyle\frac{1}{\displaystyle\alpha_{3}}
  \end{bmatrix}.
\end{align}$$

where $\displaystyle\alpha_{3} = \displaystyle\frac{n_{A}n_{C}}{n_{B}}$.
Therefore, the propagation of light through such a structure of repeating ABC layers is represented by the following matrix equation,

$$\begin{align}
  \begin{bmatrix}
          E_{0}^{(+)}\\
          E_{0}^{(-)}
  \end{bmatrix} = J_{1(3)}J_{2(3)}^{r-1}M_CJ_{3(3)}^{r-1}J_{4(3)}
  \begin{bmatrix}
          E_{N+1}^{(+)}\\
          0
  \end{bmatrix},
\end{align}$$

where the $M_C$ represents the matching matrix of the center layer, $E_{0}^{(+)}$, $E_{0}^{(-)}$ and $E_{N+1}^{(+)}$ are the incident, reflected and transmitted electric fields of the structure.
    
The transfer matrices for the ABCD unit layer are given as follows:

$$\begin{align}
  &J_{1 (4)} = \displaystyle\frac{1}{2}
  \begin{bmatrix}
          \displaystyle\frac{1}{\displaystyle\alpha_{4}} + \displaystyle\alpha_{4}n_{D} & \displaystyle\frac{1}{\displaystyle\alpha_{4}} - \displaystyle\alpha_{4}n_{D}\\
          \displaystyle\frac{1}{\displaystyle\alpha_{4}} - \displaystyle\alpha_{4}n_{D} & \displaystyle\frac{1}{\displaystyle\alpha_{4}} + \displaystyle\alpha_{4}n_{D}
  \end{bmatrix},
\end{align}$$

$$\begin{align}
  &J_{2 (4)} = \displaystyle\frac{1}{2}
  \begin{bmatrix}
          \displaystyle\frac{1}{\displaystyle\alpha_{4}} + \displaystyle\alpha_{4} & \displaystyle\frac{1}{\displaystyle\alpha_{4}} - \displaystyle\alpha_{4}\\
          \displaystyle\frac{1}{\displaystyle\alpha_{4}} - \displaystyle\alpha_{4} & \displaystyle\frac{1}{\displaystyle\alpha_{4}} + \displaystyle\alpha_{4}
  \end{bmatrix},
\end{align}$$

$$\begin{align}
  &J_{3 (4)} = \displaystyle\frac{1}{2}
  \begin{bmatrix}
          \displaystyle\frac{1}{\displaystyle\alpha_{4}} + \displaystyle\alpha_{4} & - \displaystyle\frac{1}{\displaystyle\alpha_{4}} + \displaystyle\alpha_{4}\\
          - \displaystyle\frac{1}{\displaystyle\alpha_{4}} + \displaystyle\alpha_{4} & \displaystyle\frac{1}{\displaystyle\alpha_{4}} + \displaystyle\alpha_{4}
  \end{bmatrix},
\end{align}$$

$$\begin{align}
  &J_{4 (4)} = \displaystyle\frac{1}{2}
  \begin{bmatrix}
          \displaystyle\frac{1}{\displaystyle\alpha_{4}n_{D}} + \displaystyle\alpha_{4} & - \displaystyle\frac{1}{\displaystyle\alpha_{4}n_{D}} + \displaystyle\alpha_{4}\\
          - \displaystyle\frac{1}{\displaystyle\alpha_{4}n_{D}} + \displaystyle\alpha_{4} & \displaystyle\frac{1}{\displaystyle\alpha_{4}n_{D}} + \displaystyle\alpha_{4}
  \end{bmatrix}.
\end{align}$$

where $\displaystyle\alpha_{4} = \displaystyle\frac{n_{A}n_{C}}{n_{B}n_{D}}$.
Therefore, the propagation of light through such a structure of repeating ABC layers is represented by the following matrix equation,

$$\begin{align}
  \begin{bmatrix}
          E_{0}^{(+)}\\
          E_{0}^{(-)}
  \end{bmatrix} = J_{1(3)}J_{2(3)}^{r-1}M_CJ_{3(3)}^{r-1}J_{4(3)}
  \begin{bmatrix}
          E_{N+1}^{(+)}\\
          0
  \end{bmatrix},
\end{align}$$

where the $M_C$ represents the matching matrix of the center layer, $E_{0}^{(+)}$, $E_{0}^{(-)}$ and $E_{N+1}^{(+)}$ are defined similarly in the case of structure of repeating ABC layers.

In the presence of graphene, the propagation of light is represented by the following matrix equation,

$$\begin{align}
  \begin{bmatrix}
          E_{0}^{(+)}\\
          E_{0}^{(-)}
  \end{bmatrix} = J_{1(3,4)}J_{2(3,4)}^{r-1}M_GJ_{3(3,4)}^{r-1}J_{4(3,4)}
  \begin{bmatrix}
          E_{N+1}^{(+)}\\
          0
  \end{bmatrix},
\end{align}$$

where the $M_G$ represents the matching matrix of the graphene layer, which is expressed as follows,

$$\begin{align}
  &M_{G} = \displaystyle\frac{1}{2}
  \begin{bmatrix}
          2 + \displaystyle\frac{Z_{0}}{n_{m}} \sigma & \displaystyle\frac{Z_{0}}{n_{m}} \sigma \\
          - \displaystyle\frac{Z_{0}}{n_{m}} \sigma & 2 - \displaystyle\frac{Z_{0}}{n_{m}} \sigma
  \end{bmatrix},
\end{align}$$

where $\sigma=e^2/4\hbar$ is the optical conductivity of undoped graphene, $n$ is refractive index of the material surrounding graphene, which is symbolized by the $m$ index. Here, we have $n=n_C$ and $n=n_D$ for the cases of ABC and ABCD structures, respectively.
In the case of no graphene, the electric field at the center of the structure $E_c=E_c^{(+)}+E_c^{(-)}$is given by the following matrix equation,

$$\begin{align}
  \begin{bmatrix}
          E_{c}^{(+)}\\
          E_{c}^{(-)}
  \end{bmatrix} = \left( J_{1(3,4)}J_{2(3,4)}^{r-1} \right)^{-1}
  \begin{bmatrix}
          E_{0}^{(+)}\\
          E_{0}^{(-)}
  \end{bmatrix}.
\end{align}$$

On the other hand, in the presence of graphene, the electric field at the center of the structure is given by the following matrix equation,

$$\begin{align}
  \begin{bmatrix}
          E_{c}^{(+)}\\
          E_{c}^{(-)}
  \end{bmatrix} = \left( J_{1(3,4)}J_{2(3,4)}^{r-1} M_G\right)^{-1}
  \begin{bmatrix}
          E_{0}^{(+)}\\
          E_{0}^{(-)}
  \end{bmatrix}
\end{align}$$

```python
# This program calculates the absorption spectra of the mirror geometry by using transfer matrix
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib as mpl
from numpy.linalg import matrix_power
from scipy.linalg import fractional_matrix_power
from numpy.linalg import inv
```

```python
# The necessary constants
hbar  = 1.055e-34            # the reduced Planck constant
c     = 299792458            # the light speed at vacuum
e     = 1.602e-19            # electron charge
sig   = e * e / (4 * hbar)   # the optical conductivity of undoped graphene
lmbd  = 632e-9               # the wavelength
omega = 2 * np.pi * c / lmbd # the angular frequency
na    = 2                    # the refractive index of B medium
nb    = 1                    # the refractive index of B medium
nc    = 2                    # the refractive index of C medium
nd    = 1
eps0  = 8.8502e-12           # the vacuum permitivity
```

```python
# The matching matrix

def M(x,y): #x is n_i+1, y is n_i
    A = 1 + x / y
    B = 1 - x / y
    return 0.5 * np.array([[A,B],[B,A]])

# The propagation matrix

P = np.array([[-1j,0],[0,1j]])

# The transfer matrix for mirror symmetry with s repetition

# ABC
def TF(x,s): # x is the alpha and s is the repetition
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P
    J2 = M(na,nc) @ P @ M(nb,na) @ P @ M(nc,nb) @ P
    J3 = P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(nc,na)
    J4 = P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(1,na)
    JT = J1 @ matrix_power(J2,s-1) @ matrix_power(J3,s-1) @ J4
    return JT

# ABCD
def TF4(x,s): # x is the alpha and s is the repetition
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J2 = M(na,nd) @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J3 = P @ M(nc,nd) @ P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(nd,na)
    J4 = P @ M(nc,nd) @ P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(1,na)
    JT = J1 @ matrix_power(J2,s-1) @ matrix_power(J3,s-1) @ J4
    return JT

# Calculation of electric field at the graphene

# ABC
def ec(x,s):# x is the alpha and s is the repetition
    et = 1/(TF(x,s)[0,0]) 
    ep = (TF(x,s)[1,0]) *et
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P
    J2 = M(na,nc) @ P @ M(nb,na) @ P @ M(nc,nb) @ P
    JT = J1 @ matrix_power(J2,s-1)
    JC = inv(JT) @ np.array([[et],[0]])
    return np.abs(JC[0] + JC[1])

# ABCD
def ec4(x,s):# x is the alpha and s is the repetition
    et = 1/(TF4(x,s)[0,0]) 
    ep = (TF4(x,s)[1,0]) *et
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J2 = M(na,nd) @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    JT = J1 @ matrix_power(J2,s-1)
    JC = inv(JT) @ np.array([[et],[0]])
    return np.abs(JC[0] + JC[1])
```

```python
# The matching matrix

def M(x,y): #x is n_i+1, y is n_i
    A = 1 + x / y
    B = 1 - x / y
    return 0.5 * np.array([[A,B],[B,A]])

# The propagation matrix

P = np.array([[-1j,0],[0,1j]])

def MG(n): #x is n_i+1, y is n_i, z is the conductivity
    A = 2 + 377 * sig / n
    B = 377 * sig / n
    C = -377 * sig / n
    D = 2 - 377 * sig / n
    return 0.5 * np.array([[A,B],[C,D]])

# The transfer matrix for mirror symmetry with s repetition

def TF(x,s): # x is the alpha and s is the repetition
    nb = na * nc / x
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P
    J2 = M(na,nc) @ P @ M(nb,na) @ P @ M(nc,nb) @ P
    J3 = P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(nc,na)
    J4 = P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(1,na)
    JT = J1 @ matrix_power(J2,s-1) @ MG(nc) @ matrix_power(J3,s-1) @ J4
    return JT

def TF4(x,s): # x is the alpha and s is the repetition
    na = 2
    nc = 2
    nbd = na * nc / x
    nb  = nbd**0.5
    nd  = nbd**0.5
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J2 = M(na,nd) @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J3 = P @ M(nc,nd) @ P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(nd,na)
    J4 = P @ M(nc,nd) @ P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(1,na)
    JT = J1 @ matrix_power(J2,s-1) @ MG(nd) @ matrix_power(J3,s-1) @ J4
    return JT


# Calculation of electric field at the graphene

#ABC

def ec(x,s):# x is the alpha and s is the repetition
    nb = na * nc / x
    et = 1/(TF(x,s)[0,0]) 
    ep = (TF(x,s)[1,0]) *et
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P
    J2 = M(na,nc) @ P @ M(nb,na) @ P @ M(nc,nb) @ P
    JT = J1 @ matrix_power(J2,s-1) @ MG(nc)
    JC = inv(JT) @ np.array([[1],[ep]])
    return np.abs(JC[0] + JC[1])

# ABCD

def ec4(x,s):# x is the alpha and s is the repetition
    na = 2
    nc = 2
    nbd = na * nc / x
    nb  = nbd**0.5
    nd  = nbd**0.5
    et = 1/(TF4(x,s)[0,0]) 
    ep = (TF4(x,s)[1,0]) *et
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J2 = M(na,nd) @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    JT = J1 @ matrix_power(J2,s-1) @ MG(nd)
    JC = inv(JT) @ np.array([[1],[ep]])
    return np.abs(JC[0] + JC[1])


# Calculation of absorption
# ABC
def ap(x,s):# x is the alpha and s is the repetition
    T = np.abs(1/(TF(x,s)[0,0]))**2
    R = np.abs((TF(x,s)[1,0])/(TF(x,s)[0,0]))**2
    return 1 - T - R

#ABCD
def ap4(x,s):# x is the alpha and s is the repetition
    T = np.abs(1/(TF4(x,s)[0,0]))**2
    R = np.abs((TF4(x,s)[1,0])/(TF4(x,s)[0,0]))**2
    return 1 - T - R


# For ABC

dat = 17
AA  = np.zeros(dat+1) # Positive alpha
NAA  = np.zeros(dat+1) # Negative alpha
EE1  = np.zeros(dat+1) # Electric Field for odd alpha
EE2 = np.zeros(dat+1) # Electrical Field for even alpha
NEE1 = np.zeros(dat+1) # Electrical Field for odd alpha (negative)
NEE2 = np.zeros(dat+1) # Electrical Field for even alpha (negative)
AB1  = np.zeros(dat+1) # Absorption for odd alpha
AB2 = np.zeros(dat+1) # Absorption for even alpha
NAB1 = np.zeros(dat+1) # Absorption for odd alpha (negative)
NAB2 = np.zeros(dat+1) # Absorption for even alpha (positive)

# Numerical (Positive alpha)
for i in range (dat+1):
    AA[i]     = (i+1)/2
    EE1[i]    = ec((i+1)/2,1)
    EE2[i]    = ec((i+1)/2,2)
    AB1[i]     = ap((i+1)/2,1)
    AB2[i]     = ap((i+1)/2,2)

# Numerical (Negative alpha)
for i in range (-18,0,1):
    NAA[i]     = i/2
    NEE1[i]    = ec(i/2,1)
    NEE2[i]    = ec(i/2,2)
    NAB1[i]     = ap(i/2,1)
    NAB2[i]     = ap(i/2,2)

# Analytical
x   = np.linspace(-4,4,9000)
y1  = 4*377*sig*x**2/(2*x**2+377*sig)**2 # Absorption for odd alpha
ye1 = np.absolute(2*x/(2*x**2+377*sig)) # Electric field for odd alpha
y2  = 4*377*sig/(2+377*sig)**2 # Absorption for even alpha
ye2 = np.absolute((2+2*377*sig)/(2+377*sig)) # Electric field for even alpha

mpl.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 24
plt.rcParams['axes.linewidth'] = 1.5

fig = plt.figure(figsize=(8, 8))
axee_ABC   = fig.add_axes([0,     0.55, 0.42, 0.45])
axabs_ABC  = fig.add_axes([0,     0,   0.42, 0.4])
axee_ABCD  = fig.add_axes([0.58,  0.55, 0.42, 0.45])
axabs_ABCD = fig.add_axes([0.58,  0,   0.42, 0.4])

axee_ABC.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axee_ABC.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axee_ABCD.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axee_ABCD.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axabs_ABC.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axabs_ABC.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axabs_ABCD.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axabs_ABCD.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')

axee_ABC.yaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axee_ABC.yaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axee_ABCD.yaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axee_ABCD.yaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axabs_ABC.yaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axabs_ABC.yaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axabs_ABCD.yaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axabs_ABCD.yaxis.set_tick_params(which='minor', size=5, direction='in', top='on')

# Plotting Electric Field 
axee_ABC.plot(x, ye1,color='C0') # Analitic for odd alpha
axee_ABC.plot([], [], '-D', color='C0', label = 'r = odd')
axee_ABC.axhline(ye2,color='C2',linestyle='--') # Analitic for even alpha
axee_ABC.plot([], [], '-x', color='C2', label = 'r = even')
# Numerical
axee_ABC.scatter(AA,EE2, color='C2',marker='x')
axee_ABC.scatter(AA,EE1, color='C0',marker='d')
axee_ABC.scatter(NAA,NEE2, color='C2',marker='x')
axee_ABC.scatter(NAA,NEE1, color='C0',marker='d')
axee_ABC.xaxis.set_major_locator(mpl.ticker.MultipleLocator(2))
axee_ABC.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
axee_ABC.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1))
axee_ABC.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
axee_ABC.set_ylabel('$|E_c/E_0^{(+)}|$',size  = 18,labelpad=18)
axee_ABC.set_xlabel(r'$\alpha_3$',size  = 18)
axee_ABC.set_xlim(-4,4)
axee_ABC.set_ylim(0,7)
axee_ABC.legend(loc='best',fontsize=12,markerscale=2,frameon=False)

# Plotting Absorption
axabs_ABC.plot(x, y1,color='C0') # Analitic for odd alpha
axabs_ABC.plot([], [], '-D', color='C0', label = 'r = odd')
axabs_ABC.axhline(y2, linestyle='--',color='C2') # Analitic for even alpha
axabs_ABC.plot([], [], '-x', color='C2', label = 'r = even')
# Numerical
axabs_ABC.scatter(AA,AB2, color='C2',marker='x')
axabs_ABC.scatter(AA,AB1, color='C0',marker='d')
axabs_ABC.scatter(NAA,NAB2, color='C2',marker='x')
axabs_ABC.scatter(NAA,NAB1, color='C0')
axabs_ABC.xaxis.set_major_locator(mpl.ticker.MultipleLocator(2))
axabs_ABC.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
axabs_ABC.yaxis.set_major_locator(mpl.ticker.MultipleLocator(0.1))
axabs_ABC.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.05))
axabs_ABC.set_ylabel('Absorbtion',size  = 18)
axabs_ABC.set_xlabel(r'$\alpha_3$',size  = 18)
axabs_ABC.set_xlim(-4,4)
axabs_ABC.set_ylim(0,0.67)
axabs_ABC.legend(loc='best',fontsize=12,markerscale=2,frameon=False)

# For ABCD
# Analytical
x   = np.linspace(-8,8,45000)
s1 = 2
s2 = 3
s3 = 5
ye1  = np.absolute(2*x**s1/(2+377*sig*x**(2*s1))) # Electric field for s1
ye2  = np.absolute(2*x**s2/(2+377*sig*x**(2*s2))) # Electric field for s2
ye3  = np.absolute(2*x**s3/(2+377*sig*x**(2*s3))) # Electric field for s3
y1   = 4*377*sig*x**(2*s1)/(2+x**(2*s1)*377*sig)**2 # Absorption for s1
y2   = 4*377*sig*x**(2*s2)/(2+x**(2*s2)*377*sig)**2 # Absorption for s2
y3   = 4*377*sig*x**(2*s3)/(2+x**(2*s3)*377*sig)**2 # Absorption for s3

# Numerical
dat = 35
AA  = np.zeros(dat+1) # Positive alpha
EE1 = np.zeros(dat+1) # Electrical field for s1
EE2 = np.zeros(dat+1) # Electrical field for s2
EE3 = np.zeros(dat+1) # Electrical field for s3
AB1 = np.zeros(dat+1) # Absorption for s1
AB2 = np.zeros(dat+1) # Absorption for s2
AB3 = np.zeros(dat+1) # Absorption for s3
NAA  = np.zeros(dat+1) # Negative alpha
NEE1 = np.zeros(dat+1) # Electrical field for s1 (negative)
NEE2 = np.zeros(dat+1) # Electrical field for s2 (negative)
NEE3 = np.zeros(dat+1) # Electrical field for s3 (negative)
NAB1 = np.zeros(dat+1) # Absorption for s1 (negative)
NAB2 = np.zeros(dat+1) # Absorption for s2 (negative)
NAB3 = np.zeros(dat+1) # Absorption for s3 (negative)

for i in range (dat+1):
    AA[i]     = (i+1)/2
    EE1[i]    = ec4((i+1)/2,s1)
    EE2[i]    = ec4((i+1)/2,s2)
    EE3[i]    = ec4((i+1)/2,s3)
    AB1[i]    = ap4((i+1)/2,s1)
    AB2[i]    = ap4((i+1)/2,s2)
    AB3[i]    = ap4((i+1)/2,s3)
    
for i in range (-1*(dat+1),0,1):
    NAA[i]     = (i)/2
    NEE1[i]    = ec4((i)/2,s1)
    NEE2[i]    = ec4((i)/2,s2)
    NEE3[i]    = ec4((i)/2,s3)
    NAB1[i]     = ap4((i)/2,s1)
    NAB2[i]     = ap4((i)/2,s2)
    NAB3[i]     = ap4((i)/2,s3)

# Plotting electric field
# Analytical
axee_ABCD.plot(x, ye1,color='C0')
axee_ABCD.plot([], [], '-^', color='C0', label = 'r = 2')
axee_ABCD.plot(x, ye2,color='red')
axee_ABCD.plot([], [], '-o', color='red', label = 'r = 3')
axee_ABCD.plot(x, ye3,color='C2')
axee_ABCD.plot([], [], '-s', color='C2', label = 'r = 5')

# Numeric
axee_ABCD.scatter(AA,EE1,color='C0',marker='^')
axee_ABCD.scatter(AA,EE2,color='red',marker='o')
axee_ABCD.scatter(AA,EE3,color='C2',marker='s')
axee_ABCD.scatter(NAA,NEE1,color='C0',marker='^')
axee_ABCD.scatter(NAA,NEE2,color='red',marker='o')
axee_ABCD.scatter(NAA,NEE3,color='C2',marker='s')
axee_ABCD.xaxis.set_major_locator(mpl.ticker.MultipleLocator(4))
axee_ABCD.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(1))
axee_ABCD.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1))
axee_ABCD.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
axee_ABCD.set_ylabel('$|E_c/E_0^{(+)}|$',size  = 18,labelpad=18)
axee_ABCD.set_xlabel(r'$\alpha_4$',size  = 18)
axee_ABCD.set_ylim(0,7)
axee_ABCD.set_xlim(-8,8)
axee_ABCD.legend(loc='best',fontsize=12,markerscale=2,frameon=False)

# Plotting absorption
# Analytical
axabs_ABCD.plot(x, y1,color='C0')
axabs_ABCD.plot([], [], '-^', color='C0', label = 'r = 2')
axabs_ABCD.plot(x, y2,color='red')
axabs_ABCD.plot([], [], '-o', color='red', label = 'r = 3')
axabs_ABCD.plot(x, y3,color='C2')
axabs_ABCD.plot([], [], '-s', color='C2', label = 'r = 5')

# Numerical
axabs_ABCD.scatter(AA,AB1, color='C0',marker='^')
axabs_ABCD.scatter(AA,AB2, color='red',marker='o')
axabs_ABCD.scatter(AA,AB3, color='C2',marker='s')
axabs_ABCD.scatter(NAA,NAB1, color='C0',marker='^')
axabs_ABCD.scatter(NAA,NAB2, color='red',marker='o')
axabs_ABCD.scatter(NAA,NAB3, color='C2',marker='s')
axabs_ABCD.xaxis.set_major_locator(mpl.ticker.MultipleLocator(4))
axabs_ABCD.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(1))
axabs_ABCD.yaxis.set_major_locator(mpl.ticker.MultipleLocator(0.1))
axabs_ABCD.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.05))
axabs_ABCD.set_ylabel('Absorption',size  = 18)
axabs_ABCD.set_xlabel(r'$\alpha_4$',size  = 18)
axabs_ABCD.set_xlim(-8,8)
axabs_ABCD.set_ylim(0,0.67)
axabs_ABCD.legend(loc='best',fontsize=12,markerscale=2,frameon=False)

axee_ABC.tick_params(axis='both', which='major', pad=13)
axee_ABCD.tick_params(axis='both', which='major', pad=13)
axabs_ABC.tick_params(axis='both', which='major', pad=13)
axabs_ABCD.tick_params(axis='both', which='major', pad=13)
axee_ABC.xaxis.label.set_fontsize(25)
axee_ABC.yaxis.label.set_fontsize(25)
axabs_ABC.xaxis.label.set_fontsize(25)
axabs_ABC.yaxis.label.set_fontsize(25)
axee_ABCD.xaxis.label.set_fontsize(25)
axee_ABCD.yaxis.label.set_fontsize(25)
axabs_ABCD.xaxis.label.set_fontsize(25)
axabs_ABCD.yaxis.label.set_fontsize(25)


axee_ABC.text(-6.75, 6.7, '(a)', fontsize=25)
axee_ABC.text(-3.5, 6.2, 'ABC', fontsize=20)
axee_ABCD.text(-13.5, 6.7, '(b)', fontsize=25)
axee_ABCD.text(-6.5, 6.2, 'ABCD', fontsize=20)
axabs_ABC.text(-6.75, 0.65, '(c)', fontsize=25)
axabs_ABC.text(-3.5, 0.59, 'ABC', fontsize=20)
axabs_ABC.text(-3.5, 0.032, '0.023', fontsize=18, color='green')
axabs_ABCD.text(-13.5, 0.65, '(d)', fontsize=25)
axabs_ABCD.text(-6.5, 0.59, 'ABCD', fontsize=20)

plt.savefig('figure4.pdf',bbox_inches='tight')
```

```python
# The matching matrix

def M(x,y): #x is n_i+1, y is n_i
    A = 1 + x / y
    B = 1 - x / y
    return 0.5 * np.array([[A,B],[B,A]])

# The propagation matrix

P = np.array([[-1j,0],[0,1j]])

def MG(n): #x is n_i+1, y is n_i, z is the conductivity
    A = 2 + 377 * sig / n
    B = 377 * sig / n
    C = -377 * sig / n
    D = 2 - 377 * sig / n
    return 0.5 * np.array([[A,B],[C,D]])

def TF4(x,s): # x is the alpha and s is the repetition
    na = 2
    nc = 2
    nbd = na * nc / x
    nb  = nbd**0.5
    nd  = nbd**0.5
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J2 = M(na,nd) @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J3 = P @ M(nc,nd) @ P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(nd,na)
    J4 = P @ M(nc,nd) @ P @ M(nb,nc) @ P @ M(na,nb) @ P @ M(1,na)
    JT = J1 @ matrix_power(J2,s-1) @ MG(nd) @ matrix_power(J3,s-1) @ J4
    return JT


# Calculation of electric field at the graphene

def ec4(x,s):# x is the alpha and s is the repetition
    na = 2
    nc = 2
    nbd = na * nc / x
    nb  = nbd**0.5
    nd  = nbd**0.5
    et = 1/(TF4(x,s)[0,0]) 
    ep = (TF4(x,s)[1,0]) *et
    J1 = M(na,1)  @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    J2 = M(na,nd) @ P @ M(nb,na) @ P @ M(nc,nb) @ P @ M(nd,nc) @ P
    JT = J1 @ matrix_power(J2,s-1) @ MG(nd)
    JC = inv(JT) @ np.array([[1],[ep]])
    return np.abs(JC[0] + JC[1])

# Calculation of absorption

def ap4(x,s):# x is the alpha and s is the repetition
    T = np.abs(1/(TF4(x,s)[0,0]))**2
    R = np.abs((TF4(x,s)[1,0])/(TF4(x,s)[0,0]))**2
    return 1 - T - R

# Analytical
s   = np.linspace(1,8,8) # Number of repetition
al1 = 2
al2 = 3
al3 = 4
ye1  = 2*al1**s/(2+377*sig*al1**(2*s)) # Electric field for alpha = 2
ye2  = 2*al2**s/(2+377*sig*al2**(2*s)) # Electric field for alpha = 3
ye3  = 2*al3**s/(2+377*sig*al3**(2*s)) # Electric field for alpha = 4
y1   = 4*377*sig*al1**(2*s)/(2+al1**(2*s)*377*sig)**2 # Absorption for alpha = 2
y2   = 4*377*sig*al2**(2*s)/(2+al2**(2*s)*377*sig)**2 # Absorption for alpha = 3
y3   = 4*377*sig*al3**(2*s)/(2+al3**(2*s)*377*sig)**2 # Absorption for alpha = 4

# Numerical
dat = 8
SS  = np.zeros(dat+1) # Number of repetition
EE2 = np.zeros(dat+1) # Electric field for alpha = 2
EE3 = np.zeros(dat+1) # Electric field for alpha = 3
EE4 = np.zeros(dat+1) # Electric field for alpha = 4
AB2 = np.zeros(dat+1) # Absorption for alpha = 2
AB3 = np.zeros(dat+1) # Absorption for alpha = 3
AB4 = np.zeros(dat+1) # Absorption for alpha = 4


for i in range(dat+1):
    SS[i]     = i
    EE2[i]    = ec4(2,i)
    EE3[i]    = ec4(3,i)
    EE4[i]    = ec4(4,i)
    AB2[i]    = ap4(2,i)
    AB3[i]    = ap4(3,i)
    AB4[i]    = ap4(4,i)

mpl.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 24
plt.rcParams['axes.linewidth'] = 1.5

fig = plt.figure(figsize=(8, 4))

axEE = fig.add_axes([0,   0, 0.42, 1])
axAB = fig.add_axes([0.58, 0, 0.42, 1])

axEE.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axEE.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axAB.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axAB.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')

axEE.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axEE.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')
axAB.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axAB.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')


# Plotting the eletric field
# Analytical
axEE.plot(s, ye1,label=r'$\alpha$ = 2',linestyle='--',color='C0',marker='^')
axEE.plot(s, ye2,label=r'$\alpha$ = 3',linestyle='--',color='red',marker='o')
axEE.plot(s, ye3,label=r'$\alpha$ = 4',linestyle='--',color='C2',marker='s')
# Numerical
axEE.scatter(SS,EE2,color='C0',marker='^',s=100)
axEE.scatter(SS,EE3,color='red',marker='o',s=100)
axEE.scatter(SS,EE4,color='C2',marker='s',s=100)
axEE.xaxis.set_major_locator(mpl.ticker.MultipleLocator(1))
axEE.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1))
axEE.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.25))
axEE.set_ylabel('$|E_c/E_0^{(+)}|$')
axEE.set_xlabel('Repetition Number')
axEE.set_xlim([1,8])
axEE.set_ylim([0,5])
axEE.legend(loc='best',fontsize=16,markerscale=2,frameon=False)

# Plotting the absorption
# Analytical
axAB.plot(s, y1, linestyle='--',color='C0',marker='^',label=r'$\alpha$ = 2')
axAB.plot(s, y2, linestyle='--',color='red',label=r'$\alpha$ = 3',marker='o')
axAB.plot(s, y3, linestyle='--',color='C2',marker='s',label=r'$\alpha$ = 4')
# Numerical
axAB.scatter(SS,AB2, color='C0',marker='^',s=100)
axAB.scatter(SS,AB3,color='red',marker='o',s=100)
axAB.scatter(SS,AB4, color='C2',marker='s',s=100)
axAB.xaxis.set_major_locator(mpl.ticker.MultipleLocator(1))
axAB.yaxis.set_major_locator(mpl.ticker.MultipleLocator(0.1))
axAB.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.025))
axAB.set_ylabel('Absorption')
axAB.set_xlabel('Repetition Number')
axAB.set_xlim(1,8)
axAB.set_ylim(0,0.52)
axAB.legend(loc='best',fontsize=16,markerscale=2,frameon=False)

axEE.tick_params(axis='both', which='major', pad=13)
axAB.tick_params(axis='both', which='major', pad=13)
axEE.xaxis.label.set_fontsize(25)
axEE.yaxis.label.set_fontsize(25)
axAB.xaxis.label.set_fontsize(25)
axAB.yaxis.label.set_fontsize(25)

axEE.text(-1, 4.75, '(a)', fontsize=25)
axAB.text(-1.5, 0.49, '(b)', fontsize=25)

plt.savefig('figure7.pdf',bbox_inches='tight')
```
